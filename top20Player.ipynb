```python
# Import the pathlib and csv library
# initial imports
import os
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import hvplot.pandas
from pathlib import Path
from dotenv import load_dotenv
import random
import numpy as np
%matplotlib inline


#Setting the different CSV paths
data_2015_path = Path("./Resources/Historic_Data/fanduel_salary_production/NFL-HistoricalProduction_2015.csv")
data_2016_path = Path("./Resources/Historic_Data/fanduel_salary_production/NFL-HistoricalProduction_2016.csv")
data_2017_path = Path("./Resources/Historic_Data/fanduel_salary_production/NFL-HistoricalProduction_2017.csv")
data_2018_path = Path("./Resources/Historic_Data/fanduel_salary_production/NFL-HistoricalProduction_2018.csv")
data_2019_path = Path("./Resources/Historic_Data/fanduel_salary_production/NFL-HistoricalProduction_2019.csv")
data_2020_path = Path("./Resources/Historic_Data/fanduel_salary_production/NFL-HistoricalProduction_2020.csv")

#Reading in the CSV into into seperate dataframes
df_1 = pd.read_csv(data_2015_path)
df_2 = pd.read_csv(data_2016_path)
df_3 = pd.read_csv(data_2017_path)
df_4 = pd.read_csv(data_2018_path)
df_5 = pd.read_csv(data_2019_path)
df_6 = pd.read_csv(data_2020_path)
```

```python
#group by player calculate average points for each player
```

```python
df_player_avg_points_1 = df_1.groupby(['Player','P']).mean() 
df_player_avg_points_1
```

```python
df_player_avg_points_2 = df_2.groupby(['Player','P']).mean() 
df_player_avg_points_2
```

```python
df_player_avg_points_3 = df_3.groupby(['Player','P']).mean() 
df_player_avg_points_3
```

```python
df_player_avg_points_4 = df_4.groupby(['Player','P']).mean() 
df_player_avg_points_4
```

```python
df_player_avg_points_5 = df_5.groupby(['Player','P']).mean() 
df_player_avg_points_5
```

```python
df_player_avg_points_6 = df_6.groupby(['Player','P']).mean() 
df_player_avg_points_6
```

```python
#Sort by 1) position then 2)average points
df_player_sorted_position_fd_avg_1 = df_player_avg_points_1.sort_values(['P', 'FD Points'], ascending=[True,False])
df_player_sorted_position_fd_avg_1
```

```python
df_player_sorted_position_fd_avg_2 = df_player_avg_points_2.sort_values(['P', 'FD Points'], ascending=[True,False])
df_player_sorted_position_fd_avg_2
```

```python
df_player_sorted_position_fd_avg_3 = df_player_avg_points_3.sort_values(['P', 'FD Points'], ascending=[True,False])
df_player_sorted_position_fd_avg_3
```

```python
df_player_sorted_position_fd_avg_4 = df_player_avg_points_4.sort_values(['P', 'FD Points'], ascending=[True,False])
df_player_sorted_position_fd_avg_4
```

```python
df_player_sorted_position_fd_avg_5 = df_player_avg_points_5.sort_values(['P', 'FD Points'], ascending=[True,False])
df_player_sorted_position_fd_avg_5
```

```python
df_player_sorted_position_fd_avg_6 = df_player_avg_points_6.sort_values(['P', 'FD Points'], ascending=[True,False])
df_player_sorted_position_fd_avg_6
```

```python
##Select top 20 players for each position (QB, RB, WR, TE) and put into a dataframe

df_player_top_20_FB_1 = df_player_sorted_position_fd_avg_1.filter(like='QB', axis=0).head(20)
df_player_top_20_WR_1 = df_player_sorted_position_fd_avg_1.filter(like='WR', axis=0).head(20)
df_player_top_20_RB_1 = df_player_sorted_position_fd_avg_1.filter(like='RB', axis=0).head(20)
df_player_top_20_TE_1 = df_player_sorted_position_fd_avg_1.filter(like='TE', axis=0).head(20)

#combined_df_player_top_20

combined_df_player_top_20_1 = pd.concat([df_player_top_20_FB_1, df_player_top_20_WR_1,df_player_top_20_RB_1,df_player_top_20_TE_1], axis=0)
combined_df_player_top_20_1
```

```python
##Create a plot allowing user to view top 20 players (position drop down)

combined_df_player_top_20_1.hvplot.scatter(x='Player',y="FD Points",ylabel='Average_Points',ylim={10,30},groupby='P', rot=90)
```

```python
##Create another plot/chart which compares positions based on average points etc.

combined_df_player_top_20_1.hvplot.bar(x='Player',y="FD Points",ylabel='Average_Points',ylim={10,30},groupby='P', rot=90)
```
